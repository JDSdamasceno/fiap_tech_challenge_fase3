# Funções de Suporte


```python
def filtrar_nulos(df, coluna):
    """
    Retorna apenas as linhas onde
    a coluna especificada possui valores nulos.
    """
    return df[df[coluna].isna()]


def preencher_nulos(df, colunas, valor_preenchimento):
    df_copia = df.copy()
    df_copia[colunas] = df_copia[colunas].fillna(valor_preenchimento)

    return df_copia


def identificar_atraso(valor):
    if valor > 0:
        return 1
    return 0
```

# Identificação:

- dados do tipo DOT e do Tipo IATA na mesma coluna,
- Tabela airpot possui apenas IATA
- Exclusão das linhas DOT da Tabela flights
- Retirado os Casos de Cancelados (dados que não contribuem para Analise de Atrasado na Partida)
- Retirado os casos de dados Redirecionados (dados que não contribuem para Analise de Atrasado na Partida)


```python
flights.shape
```

```python
flights_limpo = flights[
    (flights["ORIGIN_AIRPORT"].str.len() == 3)
    & (flights["DESTINATION_AIRPORT"].str.len() == 3)
]
```

```python
flights_limpo.shape
```

```python
flights_limpo = flights_limpo[
    (flights_limpo["CANCELLED"] == 0) & (flights_limpo["DIVERTED"] == 0)
]
```

```python
print(f"Linhas antes: {len(flights)}")
print(f"Linhas depois: {len(flights_limpo)}")
print(f"Removidas: {len(flights) - len(flights_limpo)}")
```

# Junção das Tabelas


```python
airlines_clean = airlines.rename(
    columns={"IATA_CODE": "AIRLINE", "AIRLINE": "AIRLINE_NAME"}
)


origin_airports = airports.rename(
    columns={
        "IATA_CODE": "ORIGIN_AIRPORT",
        "AIRPORT": "ORIGIN_AIRPORT_NAME",
        "CITY": "ORIGIN_CITY",
        "STATE": "ORIGIN_STATE",
        "COUNTRY": "ORIGIN_COUNTRY",
        "LATITUDE": "ORIGIN_LATITUDE",
        "LONGITUDE": "ORIGIN_LONGITUDE",
    }
)

dest_airports = airports.rename(
    columns={
        "IATA_CODE": "DESTINATION_AIRPORT",
        "AIRPORT": "DESTINATION_AIRPORT_NAME",
        "CITY": "DESTINATION_CITY",
        "STATE": "DESTINATION_STATE",
        "COUNTRY": "DESTINATION_COUNTRY",
        "LATITUDE": "DESTINATION_LATITUDE",
        "LONGITUDE": "DESTINATION_LONGITUDE",
    }
)

flights_merged = (
    flights_limpo.merge(airlines_clean, on="AIRLINE", how="left")
    .merge(origin_airports, on="ORIGIN_AIRPORT", how="left")
    .merge(dest_airports, on="DESTINATION_AIRPORT", how="left")
)
```

# criando a Coluna Target

- Coluna que será usada principalmente para Modelo de Classificação


```python
flights_merged["DELAY"] = flights_merged["DEPARTURE_DELAY"].apply(
    identificar_atraso
)
```
